```python
import ast
```

```python
def crew_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return (l)
```

```python
movies['crew']=movies['crew'].apply(crew_director)
```

```python
def process1(obj):
    y=[]
    j=0
    for i in ast.literal_eval(obj):
        if j!=3:
            y.append(i['name'])
            j=j+1
        else:
            break

    return(y)
```

```python
def process2(obj):
    y=[]
    for i in ast.literal_eval(obj):
        y.append(i['name'])
    return (y)
```

```python
movies['genres']=movies['genres'].apply(process2)
movies['cast']=movies['cast'].apply(process1)

#movies['keywords']=movies['keywords'].apply(process2)
```

```python
def crew_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i('job')==i('Director'):
            l.append(i['name'])
            break
    return (l)
```

```python
movies['keywords']=movies['keywords'].apply(process2)
```

```python
movies.head(5)
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())


        
```

```python
movies['tags']=movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['crew']
```

```python
def removeSpace(obj):
    y=[]
    for i in obj:
        y.append(i.replace(" ",""))
    return (y)
```

```python
movies['cast']=movies['cast'].apply(removeSpace)
```

```python
movies['crew']=movies['crew'].apply(removeSpace)
movies['genres']=movies['genres'].apply(removeSpace)
movies['keywords']=movies['keywords'].apply(removeSpace)
```

```python
movies.drop(columns=['overview','keywords','crew','cast','genres'],inplace=True)
```

```python
movies['tags']=movies['tags'].apply(lambda x: " ".join(x))
```

```python
from nltk import PorterStemmer
ps=PorterStemmer()
```

```python
def steem(text):
    l=[]
    
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(i)
```

```python
movies['tags'].apply(steem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer(max_features=5000, stop_words='english')
vec=cv.fit_transform(movies['tags']).toarray()
```

```python
vec.shape
```

```python
similarity=cosine_similarity(vec)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True)
```

```python
def reccomend(film):
    mov_idx=movies[movies['title']==film].index[0]
    dist=similarity[mov_idx]
    l=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]

    for i in l:
        print(movies.iloc[i[0]].title)
```

```python
reccomend("Avatar")
```
